```python
import pandas as pd
import numpy as np
import nltk
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn import naive_bayes
from sklearn.metrics import roc_auc_score,accuracy_score
import pickle
```

```python
nltk.download("stopwords")
```

```python
dataset = pd.read_csv('reviews.txt',sep = '\t', names =['Reviews','Comments'])
```

```python
dataset
```

```python
stopset = stopwords.words('english')
```

```python
vectorizer = TfidfVectorizer(use_idf = True,lowercase = True, strip_accents='ascii',stop_words=stopset)
```

```python
X = vectorizer.fit_transform(dataset.Comments)
y = dataset.Reviews
pickle.dump(vectorizer, open('tranform.pkl', 'wb'))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
clf = naive_bayes.svm()
clf.fit(X_train,y_train)
```

Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming you have a list of accuracy values from your trained model
accuracy_values = [0.85, 0.88, 0.92, 0.91, 0.89]

# Plotting the accuracy values
plt.figure(figsize=(10, 6))
sns.lineplot(x=range(1, len(accuracy_values) + 1), y=accuracy_values, marker='o')
plt.title('Accuracy of Trained Model Over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.xticks(range(1, len(accuracy_values) + 1))
plt.grid(True)
plt.show()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Showing the graph
plt.show()
```

```python
clf = naive_bayes.MultinomialNB()
clf.fit(X_train,y_train)
```

```python
accuracy_score(y_test,clf.predict(X_test))*100
```

```python
clf = naive_bayes.MultinomialNB()
clf.fit(X,y)
```

```python
accuracy_score(y_test,clf.predict(X_test))*100
```
